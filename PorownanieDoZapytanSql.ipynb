```python
import numpy as np
import pandas as pd
```

```python
url = ('http://storage.googleapis.com/esmartdata-courses-files/ds-bootcamp/online_retail.xlsx')
retail_raw = pd.read_excel(url)
retail_raw.head()
```

```python
retail = retail_raw.copy()
retail.info()
```

```python
retail.describe()
```

```python
retail.isnull().sum()
```

```python
retail = retail.dropna()
retail.isnull().sum()
```

```python
retail = retail[retail.Quantity >=0]
retail.info()
```

```python
retail.CustomerID = retail.CustomerID.apply(lambda x: str(int(x)))
retail.info()
```

```python
retail.head()
```

```python
# select * FROM retial;
retail
```

```python
# select quantity unitprice custumerid from retail
retail[['Quantity', 'UnitPrice', 'CustomerID']]
```

```python
# select quantity unitprice custumerid from retail limit 10
retail[['Quantity', 'UnitPrice', 'CustomerID']][:10]
```

```python
# select * from retaial where custumerid='17850'
retail[retail.CustomerID == '17850']
```

```python
retail.query('CustomerID == "17850"')
```

```python
# select from retial where custumerid-'17850' and unitprice>5
retail[(retail.CustomerID == '17850') & (retail.UnitPrice > 5)]
```

```python
#select from retial where custumerid-'17850' or country='france'
retail[(retail.CustomerID == '17850') | (retail.Country=='France')]
```

```python
# SELECT * FROM RETAIL WHERE INvoiceNo is not null
retail[retail.InvoiceNo.notnull()]
```

```python
# select customerid, count(*) from retail grup by customer id
retail.groupby('CustomerID').size()
```

```python
retail['renevue'] = retail.Quantity * retail.UnitPrice
retail.head(3)
```

```python
# select cusoemrid avg(revenue), count(*) from retial group by customerid
retail.groupby('CustomerID').aggregate({'renevue': np.mean, 'CustomerID': np.size})
```

```python
retail.groupby('CustomerID').aggregate({'renevue': np.mean, 'CustomerID': np.size}).rename(columns={'renevue': 'RevenueAverage', 'CustomerID': 'CustomerIDCount'})
```

```python
retail['InvoiceDateDay'] = retail.InvoiceDate.dt.day
retail.head()
```

```python
# select InvoiceDataDay, sum(revenue) FROM retail Group by invoiceDataDay
result = retail.groupby('InvoiceDateDay').aggregate({'renevue': np.sum})
result
```

```python
import seaborn as sns
sns.set()

result.plot(kind='bar')
```
