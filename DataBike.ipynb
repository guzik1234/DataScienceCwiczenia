```python
import pandas as pd
import pandas_datareader.data as web


df = web.DataReader(name='META', data_source='stooq')
df.head()
```

```python
df.to_csv('meta.csv')
```

```python
df_nov = df[(df.index.month==11) & (df.index.year==2021)]
df_nov
```

```python
df_nov.to_csv('fb_nov.csv')
```

```python
new_df = pd.read_csv('fb_nov.csv', index_col=0)
new_df
```

```python
df_nov.to_excel('fb_nov.xlsx')
```
