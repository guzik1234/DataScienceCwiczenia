```python
np.random.seed(0)
s = pd.Series(np.random.randn(20))
```

```python
minimum = s.aggregate('min')
minimum
```

```python
maximum = s.aggregate('max')
maximum
```

```python
suma = s.aggregate('sum')
suma
```

```python
mean = s.aggregate('mean')
mean
```

```python
std = s.aggregate('std')
std
```

```python
stats = s.aggregate(['min', 'max', 'sum', 'mean', 'std'])
stats
```

```python
s_norm = s.apply(lambda x: (x - np.mean(s)) / np.std(s))
s_norm
```

```python
sigmoid = s_norm.apply(lambda x: 1 / (1 + np.exp(-x)))
sigmoid
```
