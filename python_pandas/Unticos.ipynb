```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('aapl_us_d.csv', index_col=0)
df.columns = ['open', 'high', 'low', 'close', 'volume']
df
```

```python
df['Average'] = (df['open'] + df['close'])/2.
df
```

```python
df['Daily_change'] = df['close'] / df['close'].shift(1) -1
df
```

```python
df = df.assign(avg =(df['open'] + df['close'])/2. )
df
```

```python
max_daily_change = df['Daily_change'].max()
max_daily_change
```

```python
min_daily_change = df['Daily_change'].min()
min_daily_change
```

```python
avg_daily_change = df['Daily_change'].mean()
avg_daily_change
```

```python
df['Daily_change'].hist(bins=100)
```

```python
df['flag'] = df['Daily_change'] > 0
df
```

```python
a = df['flag'].sum()
a
```

```python
days_with_positive_change = df['flag'].sum() / df['flag'].count()
days_with_positive_change
```
