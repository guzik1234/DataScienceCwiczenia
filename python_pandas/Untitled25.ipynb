```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data=[12,34,23])
print(df)
```

```python
df = pd.DataFrame(data=[[12,53,32],[54,23,65]], index=['first', 'second'], columns=['col1', 'col2', 'col3'])
print(df)
```

```python
df = pd.DataFrame(data=[[12,53,32],[54,23,65], [325,254,333]], index=['first', 'second', 'trird'], columns=['col1', 'col2', 'col3'])
print(df)
```

```python
d = {'one': pd.Series([1,2,3]), 'two': pd.Series([4,5,6]), 'flag': ['yes','no', 'no']}
df = pd.DataFrame(d)
print(df)
```

```python
d = {'one': pd.Series([1,2,3,7,8]), 'two': pd.Series([4,5,6]) }
df = pd.DataFrame(d)
print(df)
```

```python
d = {'one': pd.Series(np.random.randn(100)),
     'two': pd.Series(np.random.randn(100)),
     'three': pd.Series(np.random.randn(100)),}
df = pd.DataFrame(d)
print(df)
```

```python
list_of_dicts = [{'apple': 2, 'orange': 3},
 {'apple': 1, 'orange': 4, 'mango': 5}]

df = pd.DataFrame(list_of_dicts)
print(df)
```
