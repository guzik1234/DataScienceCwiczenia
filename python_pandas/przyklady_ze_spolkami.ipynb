```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'amzn_us_d.csv'

df = pd.read_csv(path, index_col=0)
df.head()
```

```python
close = df['Zamkniecie']
close.head()
```

```python
close = close['2010-01-01':]
close.plot(logy=True, title='Amazon stock price' )
```

```python
plt.savefig('close.png', format='png')
```

```python
close.to_csv('close.csv', header=['close'])
```

```python
list_of_wig20_stock = ['Alior', 'CCC', 'CD Projekt']
wig20 = pd.Series(list_of_wig20_stock, name='wig20')

wig20_names = wig20.apply(lambda word: word.upper())
wig20_names
```

```python
wig20_names.isin(['CCC'])
```

```python
wig20_names[wig20_names.isin(['CCC', 'ALIOR'])]
```

```python
for company in wig20_names:
    print(company)
```

```python
for company in wig20_names:
    company = company + '_PLN'
    print(company)
```

```python
stock_with_len3 = []
for company in wig20_names:
    if len(company)==3:
      stock_with_len3.append(company)
stock_with_len3
```

```python
stock_with_len32 = [company for company in wig20_names if len(company)==3]
stock_with_len32
```

```python
stocks_with_P = [company for company in wig20_names if company.startswith('P')]
stocks_with_P
stocks_replace = [company.replace(' ', '_') for company in wig20_names]
stocks_replace
stocks_small = [company.lower() for company in stocks_replace]
stocks_small
```
