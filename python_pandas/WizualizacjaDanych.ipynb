```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range('2019-01-01', periods=1000))
```

```python
ts
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
ts = ts.cummin()
ts
ts.plot()
```

```python
ts = ts.cummax()
ts
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 5),
               index=pd.date_range('2019-01-01', periods=1000),
                  columns=list('ABCDE'))
```

```python
df = df.cumsum()
df
df.plot()
```

```python
df['B'].plot()
```

```python
abs(df.iloc[5]).plot(kind='bar', title='bar chart')
```

```python
df = pd.DataFrame(np.random.randn(100, 4),
              columns=list('ABCD'))
```

```python
df= df.cumsum()
bar_data = df.iloc[-1].apply(abs)
bar_data.plot(kind='bar', title='random generated data', color='green', alpha=0.3)
```

```python
df = pd.DataFrame({'mu_0_sigma1': np.random.randn(10000),
                   'mu_1_sigma_2': 2 * np.random.randn(10000) +1,
                   'mu_8_sigma_3': 3 * np.random.randn(10000) + 8})
df.plot.hist(bins=100, cmap='viridis', title='Differen normaln discribuion', alpha=0.5)
```

```python
df['mu_8_sigma_3'].plot.hist(bins=1000, cumulative=True, color='green', alpha=0.5)
```
