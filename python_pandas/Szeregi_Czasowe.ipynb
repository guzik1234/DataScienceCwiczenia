```python
import pandas as pd
import numpy as np
import pytz
```

```python
times_zone = list(pytz.all_timezones)
times_zone
```

```python
europe = [tz for tz in times_zone if tz.startswith('Europe')]
europe
```

```python
rng = pd.date_range('2019-01-01', periods=100, tz='Europe/Warsaw', freq='MS')
rng
```

```python
us = rng.tz_convert(tz='US/Central')
us
```

```python
rng = pd.date_range('2019-01-01', periods=100, tz='Europe/Warsaw', freq='MIN')
ts = pd.Series(np.random.rand(100), index=rng)
ts_period = ts.to_period()
ts_period.plot()
```

```python
ts_sample = ts.resample('5MIN').sum()
ts_sample
```

```python
ts_sample.plot()
```
