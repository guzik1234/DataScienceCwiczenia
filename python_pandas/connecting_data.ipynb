```python
ten_pl = pd.read_csv('ten.csv', index_col=0)
plw_pl = pd.read_csv('plw.csv', index_col=0)

ten = ten_pl.copy()
plw = plw_pl.copy()
```

```python
ten = ten_pl.copy()
plw = plw_pl.copy()
ten.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
plw.columns = ['Open', 'High', 'Low', 'Close', 'Volume']
```

```python
pd.merge(ten, plw, left_index=True, right_index=True)
```

```python
pd.merge(ten, plw, left_index=True, right_index=True, suffixes=('_ten', '_plw'))
```

```python
pd.merge(ten, plw, left_index=True, right_index=True, suffixes=('_ten', '_plw'), how='outer')
```

```python
ten['Change'] = ten['Close'] / ten['Close'].shift(1) - 1
plw['Change'] = plw['Close'] / plw['Close'].shift(1) - 1
```

```python
ten['Flag'] = ten['Change'] > 0
plw['Flag'] = plw['Change'] > 0
```

```python
ten['Flag'] = ten['Flag'].apply(int)
```
