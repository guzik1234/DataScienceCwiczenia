```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.rand(10,5),
                  columns=['A','B','C','D','E'])
df.query('(A<B)')
```

```python
df.query('(A<B) & (B<C)')
```

```python
df.query('(A<B) | (B<C)')
```

```python
df = df.round(1)

df.query('C ==[0.5, 0.3]')
```

```python
df.query('C !=0.5')
```

```python
df.query('[0.5, 0.8] in C')
```
