```python
s = pd.Series(['#sport#good#time', '#work#out#free#time', '#summer#holiday#hot'], name='hashtag')
```

```python
s = s.str.split('#')
s
```

```python
hash1 = s.str.get(1)
hash1
```

```python
hash2 = s.str.get(2)
hash2
```

```python
hash3 = s.str.get(3)
hash3
```

```python
df_concat = pd.concat([hash1, hash2, hash3], axis=0, ignore_index=True)
df_concat
```

```python
string = df_concat.str.cat(sep=' ')
string
```

```python
df_concat = pd.concat([hash1, hash2, hash3], axis=1, ignore_index=True)
df_concat
```
