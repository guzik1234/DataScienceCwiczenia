```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
fig = plt.figure()
fig.show()
```

```python
_ = plt.plot([0,1], [0,3])
```

```python
_ = plt.plot([0,1], [0,3], label = 'line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('title')
plt.legend()
```

```python
plt.plot([0,1,2], [0,3,2], label = 'line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('title')
plt.legend()
```

```python
x = np.arange(-3,3,0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)
plt.plot(x, x_sin, label = 'sin')
plt.plot(x, x_cos, label = 'cos')
_ = plt.legend()
```

```python
plt.subplot(121)
plt.plot([0,1,2], [3,2,5])
plt.subplot(122)
plt.plot([0,1,2], [3,2,4], 'r--')
```

```python
fig = plt.figure(figsize=(12,8))
fig.show()

plt.subplot(131)
plt.plot([0,1,2], [3,2,5])
plt.subplot(132)
plt.plot([0,1,2], [3,2,4], 'r--')
plt.subplot(133)
plt.plot([0,1,2], [3,2,4], 'ro')

```

```python
plt.bar(x = [0,1,2,3], height=[1.2,3.2,1.2, 5.2])
```

```python
plt.bar(x = ['a','b','c','d'], height=[1.2,3.2,1.2, 5.2])
```

```python
plt.barh(y = ['a','b','c','d'], width=[1.2,3.2,1.2, 5.2])
```
